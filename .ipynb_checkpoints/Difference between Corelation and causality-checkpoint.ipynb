```python
# demonstrates the difference between correlation and causation in a machine learning context
#The idea is to show that two variables can be correlated without having a causal relationship.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression

# Set random seed for reproducibility
np.random.seed(42)

# Simulate data
days = 365  # One year

# Temperature follows a sinusoidal wave over the year (simplified)
days_array = np.arange(days)   #This creates an array of integers from 0 to 364, representing each day in a year.
temperature = 20 + 10 * np.sin(2 * np.pi * days_array / 365) + np.random.normal(0, 1, days) 
#np.sin(...) creates a wave pattern (like a sine wave).
#2 * np.pi * days_array / 365 makes that wave repeat once per year (like seasons).
#10 * sin(...) sets the amplitude to 10  so temperature varies 10C around the average.(operates on day arrays)
#20 + ... makes the average temperature 20C.

# Ice cream sales increase with temperature
ice_cream_sales = 50 + 5 * temperature + np.random.normal(0, 5, days)
#Base sales: 50
#Increase per C: 5
#Random noise added to simulate real-world variation
#Higher temperature  more ice cream sales



# Drowning incidents also increase with temperature (people swim more)
drowning_incidents = 1 + 0.3 * temperature + np.random.normal(0, 0.5, days)
#base drowning: 1 incident
#For each 1C increase, drowning risk increases slightly.
#Noise added for realism.
#Again, temperature is the cause, not ice cream sales.


# Create DataFrame
df = pd.DataFrame({
    "Day": days_array,
    "Temperature": temperature,
    "IceCreamSales": ice_cream_sales,
    "DrowningIncidents": drowning_incidents
})
#Store all the simulated values in a pandas DataFrame to make it easier to analyze and plot.

# Show first few rows
print(df.head())

# Correlation matrix
#Computes Pearson correlation between each pair of variables.
#Helps you see how strongly variables are related.

corr = df[["IceCreamSales", "DrowningIncidents", "Temperature"]].corr()
print("\nCorrelation Matrix:")
print(corr)

#                   IceCreamSales  DrowningIncidents  Temperature
#IceCreamSales             1.000              0.84         0.91
#DrowningIncidents         0.84               1.000        0.86
#Temperature               0.91               0.86         1.000


# Visualize the correlation
sns.pairplot(df[["IceCreamSales", "DrowningIncidents", "Temperature"]])
plt.suptitle("Correlation Between Variables (But Not Causation!)", y=1.02)
plt.show()

# Scatter plot of Ice Cream Sales vs Drowning Incidents
plt.figure(figsize=(8, 5))
sns.scatterplot(x="IceCreamSales", y="DrowningIncidents", data=df)
plt.title("Correlation  Causation: Ice Cream vs Drowning")
plt.xlabel("Ice Cream Sales")
plt.ylabel("Drowning Incidents")
plt.grid(True)
plt.show()

```
